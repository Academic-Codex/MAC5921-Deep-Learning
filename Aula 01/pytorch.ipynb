```python
# Crie o tensor uma única vez
rand_tensor = torch.rand(2, 2)
print("Tensor original:")
print(rand_tensor)

# Aplique a primeira operação in-place
rand_tensor.log2_()
print("\nApós a primeira aplicação de log2_:")
print(rand_tensor)

# Aplique a segunda operação in-place
rand_tensor.log2_()
print("\nApós a segunda aplicação de log2_:")
print(rand_tensor)
```
